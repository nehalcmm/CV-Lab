```python
#Q2

import pandas as pd
import numpy as np

df = pd.read_csv("bayes.csv")

def bayesTheorem(pA,pB,pBA):
    return pA * pBA/pB

x = df['Outlook'].values.reshape(-1,1)
Y = df['Play'].values.reshape(-1,1)

yesCnt = 0
noCnt = 0
totalOutcomes = len(Y)

for i in range(0,totalOutcomes):
    if Y[i] == 'Yes':
        yesCnt += 1
    else:
        noCnt += 1

pPlay = yesCnt/totalOutcomes
pNoPlay = noCnt/totalOutcomes

yesOverCnt = 0
noOverCnt = 0
yesRainyCnt = 0
noRainyCnt = 0
yesSunnyCnt = 0
noSunnyCnt = 0

for i in range(0,totalOutcomes):
    if x[i] == 'Overcast':
        if Y[i] == 'Yes':
            yesOverCnt += 1
        else:
            noOverCnt += 1
    elif x[i] == 'Rainy':
        if Y[i] == 'Yes':
            yesRainyCnt += 1
        else:
            noRainyCnt += 1
    else:
        if Y[i] == 'Yes':
            yesSunnyCnt += 1
        else:
            noSunnyCnt += 1

pPlayOvercast = (yesOverCnt)/yesCnt
pNoPlayOvercast = (noOverCnt)/noCnt

pPlayRainy = (yesRainyCnt)/yesCnt
pNoPlayRainy = (noRainyCnt)/noCnt

pPlaySunny = (yesSunnyCnt)/yesCnt
pNoPlaySunny = (noSunnyCnt)/noCnt

pOvercast = (yesOverCnt + noOverCnt)/totalOutcomes
pRainy = (yesRainyCnt + noRainyCnt)/totalOutcomes
pSunny = (yesSunnyCnt + noSunnyCnt)/totalOutcomes

pSunnyPlay = bayesTheorem(pPlay,pSunny,pPlaySunny)
pSunnyNoPlay = bayesTheorem(pNoPlay,pSunny,pNoPlaySunny)

if pSunnyPlay > pSunnyNoPlay:
    print("If the weather is sunny, player should play since posterior probability is: ",pSunnyPlay)
else:
    print("If the weather is sunny,player should not play")
    
true_positive = df[(df['Outlook'] == 'Sunny') & (df['Play'] == 'Yes')].shape[0]
false_positive = df[(df['Outlook'] == 'Sunny') & (df['Play'] == 'No')].shape[0]
true_negative = df[(df['Outlook'] != 'Sunny') & (df['Play'] == 'No')].shape[0]
false_negative = df[(df['Outlook'] != 'Sunny') & (df['Play'] == 'Yes')].shape[0]

accuracy = (true_positive + true_negative) / totalOutcomes
precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) != 0 else 0
recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) != 0 else 0

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)

    
```
